```python
# 7 A student scored  70 marks in english ,90 marks in science 80 marks in maths and 74 in
#history.Write a python program to plot simple Bar chart of subject and calculate the average of marks

import matplotlib.pyplot as plt

# Student scores
subjects = ['English', 'Science', 'Maths', 'History']
marks = [70, 90, 80, 74]

# Plotting bar chart
plt.bar(subjects, marks, color='skyblue')
plt.title("Student's Marks in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100
plt.show()

# Calculating average
average_marks = sum(marks) / len(marks)
print("Average Marks:", average_marks)
```

```python
#

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate employee data
data = {
    'EMPID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
    'Age': np.random.randint(22, 60, 10),  # Random age between 22 and 60
    'Sales': np.random.randint(500, 1000, 10),  # Random sales between 500 and 1000
    'BMI': ['Normal', 'Overweight', 'Normal', 'Underweight', 'Normal', 'Overweight', 'Normal', 'Normal', 'Underweight', 'Overweight'],
    'Income': np.random.randint(30000, 70000, 10)  # Random income between 30,000 and 70,000
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# --------------------------------------------
# Plotting histograms for Age, Sales, and Income
plt.figure(figsize=(15, 5))

# Plot Age histogram
plt.subplot(1, 3, 1)
plt.hist(df['Age'], bins=6, color='lightblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Plot Sales histogram
plt.subplot(1, 3, 2)
plt.hist(df['Sales'], bins=5, color='lightgreen', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')

# Plot Income histogram
plt.subplot(1, 3, 3)
plt.hist(df['Income'], bins=5, color='lightcoral', edgecolor='black')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')

# Show all histograms
plt.tight_layout()
plt.show()
```

```python
# 8.  use a(Tips) given data set and plot scatter chart tips VS day and display the top 10 records
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset('tips')

# Display the top 10 records
print(tips.head(10))

# Plotting a scatter chart for Tips vs Day
plt.figure(figsize=(8, 6))
sns.scatterplot(x='day', y='tip', data=tips, color='blue', marker='o')

# Adding title and labels
plt.title('Scatter Plot of Tips vs Day', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Tip Amount', fontsize=12)

# Display the plot
plt.show()
```

```python
#9.  use aTips) given data set and plot histogram tips VS day and display the top 10 records

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset('tips')

# Display the top 10 records
print(tips.head(10))

# Plotting a histogram for Tips vs Day
plt.figure(figsize=(8, 6))
sns.histplot(data=tips, x='day', weights='tip', discrete=True, color='lightblue', kde=False)

# Adding title and labels
plt.title('Histogram of Tips vs Day', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Tips', fontsize=12)

# Display the plot
plt.show()
```

```python
# 10. use a given data set and plot bar chart tips VS day and display the top 10 records
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset('tips')

# Display the top 10 records
print(tips.head(10))

# Plotting a bar chart for Tips vs Day
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='tip', data=tips, palette='Blues_d')

# Adding title and labels
plt.title('Bar Chart of Tips vs Day', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Tip Amount', fontsize=12)

# Display the plot
plt.show()
```

```python
# 11. use a given data set and plot box chart tips VS day and display the top 10 records
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset('tips')

# Display the top 10 records
print(tips.head(10))

# Plotting a box plot for Tips vs Day
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=tips, palette='Set2')

# Adding title and labels
plt.title('Box Plot of Tips vs Day', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Tip Amount', fontsize=12)

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the employee DataFrame
data = {
    'EMPID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Age': [25, 32, 28, 45, 38, 40, 29, 35, 50, 42],
    'Sales': [200, 250, 230, 300, 280, 270, 220, 310, 260, 290],
    'BMI': ['Normal', 'Overweight', 'Normal', 'Underweight', 'Normal', 'Overweight', 'Normal', 'Normal', 'Overweight', 'Normal'],
    'Income': [50000, 60000, 55000, 65000, 70000, 75000, 52000, 68000, 59000, 62000]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Employee DataFrame:")
print(df)

# 13. Display a Pie Chart for BMI distribution
plt.figure(figsize=(8, 6))
bmi_counts = df['BMI'].value_counts()
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('BMI Distribution in the Employee Dataset', fontsize=14)
plt.show()

# 14. Display a Box Plot for Income
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Income', data=df, palette='Set1')
plt.title('Box Plot of Income by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.show()
```
